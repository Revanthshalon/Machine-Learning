# Example
- Let us take a linear function of $y=4x-2$

```
def myfunc(x):
  return (4*x)-2
```

```
import matplotlib.pyplot as plt
from matplotlib import style
style.use('fivethirtyeight')
%matplotlib inline
```

```
y=[]
for i in range (1,100):
  y.append(myfunc(i))
plt.plot(range(1,100),y,label=r'$f(x)=4x-2$')
plt.legend(loc='lower right')
plt.show()
```

**Observation:**
- The function $y = 4x-2$ is a linear function

```
def newfunc(a,b):
  return (4*(a+b)-2),((4*a)-2)+((4*b)-2)
```

```
y = {}
for i in range(1,100):
  quotient = (i//3)*3
  remainder = i%3
  y[i]={}
  y[i]['f(a+b)'],y[i]['f(a)+f(b)'] = newfunc(quotient,remainder)
  y[i]['a']=quotient
  y[i]['b']=remainder

fig, (ax1,ax2) = plt.subplots(1,2,figsize=(20,10))
ax1.plot(range(1,100),[y[i]['f(a+b)'] for i in range(1,100)],label='f(a+b)',lw=2)
ax1.plot(range(1,100), [y[i]['f(a)+f(b)'] for i in range(1,100)],label='f(a)+f(b)',lw=2)
ax2.plot(range(1,100),[y[i]['f(a+b)'] for i in range(1,100)],label='f(a+b)',lw=2)
ax2.plot(range(1,100), [y[i]['f(a)+f(b)'] for i in range(1,100)],label='f(a)+f(b)',lw=2)
ax2.set_xlim(0,50)
ax2.set_ylim(0,50)
plt.legend(loc='upper right')
plt.show()
```

```
print(f"for a linear function y=4x-2")
for i in range(1,5):
  print(f'{i}:{{\n"f(a+b):"{y[i]["f(a+b)"]},\n"f(a)+f(b)":{y[i]["f(a)+f(b)"]},\n"a:"{y[i]["a"]},\n"b:"{y[i]["b"]}\n}}')
```

**Observation:**
- We can clearly see that for the linear function $y=4x-2$,
  - for each value for ***a,b***, $\rightarrow f(a+b)\ne f(a)+f(b)$
- So we cannot use this to prove linearity of a function.
